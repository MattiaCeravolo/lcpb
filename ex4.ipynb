```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
file_name = "data_t-SNE_310101_d5_R100_e1_N800.dat"
X = np.loadtxt(file_name,dtype="float")
print(X)
print(X.shape)
```

```python
N = X.shape[0]
N0 = int(0.1*N)
N1 = int(0.3*N)
N2 = int(0.6*N)
y = np.zeros((N,))
y[:N0] = 0
y[N0:N0+N1] = 1
y[N0+N1:N] = 2
```

```python
perplexities = [5, 30, 50, 100]
fig, ax = plt.subplots(nrows=1,ncols=4,figsize=(40,10))
for i in range(4) :
    X_embedded = TSNE(n_components=2, perplexity=perplexities[i], random_state=1234).fit_transform(X)
    ax[i].scatter(X_embedded[:,0][:N0],X_embedded[:,1][:N0]);
    ax[i].scatter(X_embedded[:,0][N0:N0+N1],X_embedded[:,1][N0:N0+N1]);
    ax[i].scatter(X_embedded[:,0][N0+N1:N],X_embedded[:,1][N0+N1:N]);
    ax[i].axis("equal");
    ax[i].set_title(f"Perplexity = {perplexities[i]}")


#X_embedded = np.zeros((4,*X.shape))
```
